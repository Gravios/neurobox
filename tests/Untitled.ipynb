```python
import sys
sys.path.insert(0, "/media/cloud/code/python/")

from pymtax.io.load_xml    import load_xml
from pymtax.io.load_binary import load_binary

import matplotlib as mpl
#mpl.use("Qt5Agg")
mpl.use("TkAgg")
import matplotlib.pyplot as plt

fname_xml = '/media/data/data/A2929-200711/A2929-200711.xml'
fname_dat = '/media/data/data/A2929-200711/A2929-200711.dat'
fname_wat = '/media/data/data/A2929-200711/temp_wh.dat'

par = load_xml(fname_xml)
print(par.acquisitionSystem.peek())

dat = load_binary(fname_dat,[0,1],par);
wat = load_binary(fname_wat,[0,1],par);
```

```python
dat = dat.transpose()
wat = wat.transpose()
```

```python
%matplotlib inline
plt.figure(num=1,figsize=(12,5))
#plt.plot(dat[100000:151100])
plt.plot(wat[100000:100100,0])
plt.plot(wat[100000:100100,1])
plt.show
```

```python
import numpy as np
from scipy import signal
from scipy.fft import fftshift
import matplotlib.pyplot as plt


fs = 2*10e3
win = ('hamming')
nps = 2**14
nlp = int(2**14*0.5)
nfft = 2**15
f, t, Sxx = signal.spectrogram(dat[15000000:20000000,0], fs,win,nps,nlp,nfft)

plt.figure(num=1,figsize=(18,5))
plt.pcolormesh(t, f[0:40], Sxx[0:40,:], shading='gouraud')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
